```python
import pickle
import pprint
import shutil

import torch
from tqdm.autonotebook import tqdm

from utils import general_utils
from pytorch3d.io import save_obj
```

```python
gen_settings = {
    "real": {"center": [-5,-5,-5], "radius": 0.5, "jitter": 0.1},
    "fake": {"center": [5,5,5], "radius": 0.1, "jitter": 0.1}
}

device = torch.device("cuda:0")
num_gen = 500
# assumed that mesh is a unit sphere, is centered at origin
sphere_path = "notebooks/sphere_642.obj"
output_dir = "data/adversarial/sphere_points_test"
# just a dummy image
img_path = "notebooks/sphere_642_trans.png"
```

```python
# generating meshes
for setting in gen_settings:
    print(setting)
    setting_dir = os.path.join(output_dir, setting+"_meshes")
    os.makedirs(setting_dir, exist_ok=True)
    for i in tqdm(range(num_gen)):
        mesh = general_utils.load_untextured_mesh(sphere_path, device)
        verts = mesh.verts_packed()
        verts = verts*gen_settings[setting]["radius"] + torch.tensor(gen_settings[setting]["center"], device=device) + torch.rand(verts.shape, device=device)*gen_settings[setting]["jitter"]
        save_obj(os.path.join(setting_dir, "{num:0{width}}_{setting}.obj".format(num=i, width=len(str(num_gen)), setting=setting)), verts, mesh.faces_packed())
```

```python
# generating images
img_dir = os.path.join(output_dir, "fake_images")
os.makedirs(img_dir, exist_ok=True)
for img_name in tqdm(["{num:0{width}}_fake.png".format(num=i, width=len(str(num_gen))) for i in range(num_gen)]):
    shutil.copyfile(img_path, os.path.join(img_dir, img_name))
```

```python
# creating dummy pose dict
mesh_names = ["{num:0{width}}_fake".format(num=i, width=len(str(num_gen))) for i in range(num_gen)]
dummy_pose_dict = {mesh_name:{"dist":1, "elev":10, "azim":8} for mesh_name in mesh_names}
dummy_pose_path = os.path.join(output_dir, "poses.p")
pickle.dump(dummy_pose_dict, open(dummy_pose_path,"wb"))
pprint.pprint(pickle.load(open(dummy_pose_path, 'rb')))
```
