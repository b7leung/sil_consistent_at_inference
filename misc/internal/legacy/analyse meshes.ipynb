```python
import glob
import os
import pprint

import torch
from tqdm import tqdm

from utils import utils, network_utils
```

```python
# this notebook will check the number of vertices for all meshes in a folder
MESH_DIR = "data/onet_chair_pix3d_dann_simplified"

gpu_num = 0
device = torch.device("cuda:"+str(gpu_num))
```

```python
mesh_paths = glob.glob(os.path.join(MESH_DIR, "*.obj"))
mesh_vertex_dict = {}
for mesh_path in tqdm(mesh_paths):
    with torch.no_grad():
        mesh = utils.load_untextured_mesh(mesh_path, device)
        num_verts = mesh.verts_packed().shape[0]
        if num_verts not in mesh_vertex_dict:
            mesh_vertex_dict[num_verts] = [mesh_path]
        else:
            mesh_vertex_dict[num_verts].append(mesh_path)
```

```python
pprint.pprint(mesh_vertex_dict)
```

```python
for vertex_count in mesh_vertex_dict:
    print("{} meshes with {} vertices".format(len(mesh_vertex_dict[vertex_count]), vertex_count))
```

```python
most_common_vertex_count = 1502
```

```python
# create a text file in the mesh directory, as a list of all mesh instances with belonging to the set of meshes with the most common vertex count
output_mesh_list_path = os.path.join(MESH_DIR, "{}.lst".format(most_common_vertex_count))
with open(output_mesh_list_path, "w") as f:
    for i, mesh_path in enumerate(mesh_vertex_dict[most_common_vertex_count]):
        instance_name = str(mesh_path).split('/')[-1].replace(".obj", "")
        if i < len(mesh_vertex_dict[most_common_vertex_count])-1:
            f.write(instance_name+"\n")
        else:
            f.write(instance_name)
```
