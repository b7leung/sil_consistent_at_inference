```python
import pprint
import pickle
from pathlib import Path
import random

import torch
from torch.nn import functional as F
import torch.optim as optim
from tqdm.autonotebook import tqdm
import pandas as pd
import numpy as np
```

```python
stds = [0, 2, 4, 6, 10, 14, 18, 24]

#class_name = "03001627"
#gt_pose_path = "../occupancy_networks/out/pytorch3d_renders/{}/generation/meshes/rec_camera_params.pt".format(class_name)
#out_dir = "data/perturbed_poses/occnet/{}".format(class_name)

class_name = "chair"
gt_pose_path = "../occupancy_networks/out/pix3d/{}/generation/meshes/rec_camera_params.pt".format(class_name)
out_dir = "data/perturbed_poses/occnet_pix3d/{}".format(class_name)

os.makedirs(out_dir, exist_ok=True)
np.random.seed(0)
random.seed(0)
```

```python
gt_poses_dict = pickle.load(open(gt_pose_path, "rb"))
for std in stds:
    perturbed_poses_dict = {}
    for instance in gt_poses_dict:
        azim_additive_noise = np.random.normal(scale=std)
        elev_additive_noise = np.random.normal(scale=std)
        perturbed_poses_dict[instance] = {"azim":gt_poses_dict[instance]["azim"]+azim_additive_noise, "elev":gt_poses_dict[instance]["elev"]+elev_additive_noise, "dist":gt_poses_dict[instance]["dist"]}
    
    pickle.dump(perturbed_poses_dict, open(os.path.join(out_dir, "rec_camera_params_perturbed_std_{:02}.pt".format(std)),"wb"))
```
