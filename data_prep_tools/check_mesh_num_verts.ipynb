```python
import pprint
from pathlib import Path

import torch
from tqdm.autonotebook import tqdm
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly

from utils import general_utils
```

```python
mesh_dir = "../occupancy_networks/out/pytorch3d_renders_training/03001627/generation/meshes"

vertex_info = []
all_obj_paths = [str(path) for path in list(Path(mesh_dir).rglob('*.obj'))]
cpu_device = torch.device("cpu")
for obj_path in tqdm(all_obj_paths):
    num_verts = general_utils.load_untextured_mesh(obj_path, cpu_device).verts_packed().shape[0]
    vertex_info.append([obj_path, num_verts])
sorted_vertex_info = sorted(vertex_info, key=lambda x : x[1])
sorted_vertex_counts = [i[1] for i in sorted_vertex_info]
```

```python
show_num = 100
print("{} \n...\n {}".format(sorted_vertex_counts[:show_num], sorted_vertex_counts[-show_num:]))
```

```python
fig = px.histogram(sorted_vertex_counts, nbins=500)
fig.show()
```
